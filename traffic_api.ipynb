```python
#customer_id = "943-572-1339"

import pandas as pd
from google.ads.googleads.client import GoogleAdsClient

def main(client, customer_id, keywords_list):
    generate_historical_metrics(client, customer_id, keywords_list)

def generate_historical_metrics(client, customer_id, keywords_list):
    googleads_service = client.get_service("GoogleAdsService")
    keyword_plan_idea_service = client.get_service("KeywordPlanIdeaService")
    
    request = client.get_type("GenerateKeywordHistoricalMetricsRequest")
    request.customer_id = customer_id
    request.keywords = keywords_list
    
    # Qui puoi impostare altri parametri come la geolocalizzazione, la lingua, ecc.

    # Esegui la chiamata API e ottieni la risposta
    response = keyword_plan_idea_service.generate_keyword_historical_metrics(request=request)
    
    # Inizializza un DataFrame per conservare i risultati
    df_results = pd.DataFrame(columns=[
        'Keyword', 'AvgMonthlySearches', 'Competition', 'CompetitionIndex',
        'LowTopOfPageBid', 'HighTopOfPageBid'
    ])
    
    # Estrai e salva i risultati nel DataFrame
    for result in response.results:
        metrics = result.keyword_metrics
        df_results = df_results.append({
            'Keyword': result.text,
            'AvgMonthlySearches': metrics.avg_monthly_searches,
            'Competition': metrics.competition,
            'CompetitionIndex': metrics.competition_index,
            'LowTopOfPageBid': metrics.low_top_of_page_bid_micros,
            'HighTopOfPageBid': metrics.high_top_of_page_bid_micros,
        }, ignore_index=True)
    
    # Salva il DataFrame in un file CSV
    df_results.to_csv('historical_keyword_metrics.csv', index=False)

# Carica il DataFrame da un file CSV (o come preferisci)
# df = pd.read_csv("path/to/your/keyword_file.csv")

# DataFrame di esempio
df = pd.DataFrame({
    'Keywords': ['pneumatici', 'gomme', 'gomme auto']
})

# Inizializza il client di Google Ads
config_path = "google-ads.yaml"  # Sostituisci con il percorso al tuo file di configurazione
client = GoogleAdsClient.load_from_storage(config_path)

# Esegui la funzione principale
# account searcus
main(client, "2218314651", df['Keywords'].tolist())  # Sostituisci con il tuo ID cliente

# account personale
#main(client, "5907664432", df['Keywords'].tolist())  # Sostituisci con il tuo ID cliente

```
